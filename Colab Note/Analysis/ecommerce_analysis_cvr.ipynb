# 전환률 기준 최소, 최대 지점 분석

최소 날짜 : 31일

최대 날짜 : 16일

```python
df_16 = df.loc[df['day'] == 16].reset_index(drop = True)
```

```python
value_16 = df_16['main_category'].value_counts()
day_16 = pd.DataFrame(value_16)
```

```python
day_16 = day_16.reset_index(drop = False)

day_16.rename(columns = {'index' : 'main_category_name', 'main_category' : 'all_value_16'}, inplace = True)
```

```python
pur_16 = df_16[df_16['event_type'] == 'purchase'].reset_index(drop = True)
```

```python
pur_value_16 = pur_16['main_category'].value_counts()
day_pur_16 = pd.DataFrame(pur_value_16)
day_pur_16 = day_pur_16.reset_index(drop = False)
day_pur_16.rename(columns = {'index' : 'main_category_name', 'main_category' : 'pur_value_16'}, inplace = True)
```

```python
category_16 = pd.merge(day_16, day_pur_16)
```

```python
category_16['cvr_16'] = (category_16['pur_value_16'] / category_16['all_value_16']) * 100

category_16 = category_16.sort_values('cvr_16', ascending = False)
category_16 = category_16.reset_index(drop = True)

category_16
```

# 활성 사용자 수 최소, 최대 지점에 대한 분석

최소 날짜 : 3일

최대 날짜 : 20일

```python
df_20 = df.loc[df['day'] == 20].reset_index(drop = True)
```

```python
value_20 = df_20['main_category'].value_counts()
day_20 = pd.DataFrame(value_20)

day_20 = day_20.reset_index(drop = False)
day_20.rename(columns = {'index' : 'main_category_name', 'main_category' : 'all_value_20'}, inplace = True)

pur_20 = df_20[df_20['event_type'] == 'purchase'].reset_index(drop = True)

pur_value_20 = pur_20['main_category'].value_counts()
day_pur_20 = pd.DataFrame(pur_value_20)
day_pur_20 = day_pur_20.reset_index(drop = False)
day_pur_20.rename(columns = {'index' : 'main_category_name', 'main_category' : 'pur_value_20'}, inplace = True)

category_20 = pd.merge(day_20, day_pur_20)

category_20['cvr_20'] = (category_20['pur_value_20'] / category_20['all_value_20']) * 100

category_20 = category_20.sort_values('cvr_20', ascending = False)
category_20 = category_20.reset_index(drop = True)

category_20
```

```python
df_3 = df.loc[df['day'] == 3].reset_index(drop = True)
```

```python
value_3 = df_3['main_category'].value_counts()
day_3 = pd.DataFrame(value_3)

day_3 = day_3.reset_index(drop = False)
day_3.rename(columns = {'index' : 'main_category_name', 'main_category' : 'all_value_3'}, inplace = True)

pur_3 = df_3[df_3['event_type'] == 'purchase'].reset_index(drop = True)

pur_value_3 = pur_3['main_category'].value_counts()
day_pur_3 = pd.DataFrame(pur_value_3)
day_pur_3 = day_pur_3.reset_index(drop = False)
day_pur_3.rename(columns = {'index' : 'main_category_name', 'main_category' : 'pur_value_3'}, inplace = True)

category_3 = pd.merge(day_3, day_pur_3)

category_3['cvr_3'] = (category_3['pur_value_3'] / category_3['all_value_3']) * 100

category_3 = category_3.sort_values('cvr_3', ascending = False)
category_3 = category_3.reset_index(drop = True)

category_3
```

```python
category_da = pd.merge(category_20, category_3)

category_da
```

```python
category_all = pd.merge(category_cvr, category_da)

category_all[['main_category_name', 'all_value_16', 'cvr_16', 'all_value_31', 'cvr_31', 'all_value_20', 'cvr_20', 'all_value_3', 'cvr_3']]
```
