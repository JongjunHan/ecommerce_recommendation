# 결론

**가설**

**1) 활성 사용자 수 상위 두 날짜와 하위 두 날짜에 대한 분석을 한다.**

- session 로그 유지 시간과 Funnel 분석을 통한 cart, purchase의 연관성은 활성 사용자 수 증가에 연관이 있을 것이다.

**2) 전환률을 기준으로 최대, 최소 값을 보이는 날짜에 대한 분석을 한다.**

- session 로그 유지 시간과 Funnel 분석을 통한 cart, purchase의 연관성은 전환률에 연관성이 있을 것이다.

**결과 해석**

* Funnel분석을 통해 얻을 수 있는 결과는 결국 cart(장바구니)의 개념은 purchase(구매)에 직접적인 연관성을 도출하기에는 무리가 있다. 이는 단순히 사용자들의 편의성을 위해 제작 된 서비스의 개념으로 구매력에 직접적 영향이 있는 UI가 아니라고 판단된다.

* user_session 유지 시간의 경우 해당 날짜의 최초 session 발견기록과 최후 session 발견기록에 대한 대조를 한 결과로서 이는 굉장히 긴 타임을 나타내는 것으로 보아 결국에는 session 분석에 있어서 유지 시간 보다는 재 접속으로 봐야 할 가능성이 높다.

* 이 분석 데이터를 통해 도출할 수 있는 결과는 결국 session 로그를 분석함에 따라 결국 이 역시 크게 사용자 수 증대 및 전환률의 향상을 통한 구매력 증진과는 연관 지어서 볼 수는 없다는 판단을 할 수 있으며, 여기에서 우리는 category를 세분화 한 결과와 연관지어 구매력을 높이는 결과를 도출 하는 것이 가장 이상적이라고 판단한다.
