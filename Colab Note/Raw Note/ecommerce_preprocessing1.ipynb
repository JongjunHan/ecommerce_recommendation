```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet('/content/drive/MyDrive/ecommerce/2019-Oct-plus.parquet')
```

```python
df.head()
```

```python
df['cart'] = np.where(df['event_type'] == 'cart', 1, 0)
df['view'] = np.where(df['event_type'] == 'view', 1, 0)
```

```python
df.head()
```

```python
df1 = df.drop(['event_time', 'user_session','day', 'hour', 'minute', 'seconds'], axis = 1)
```

```python
df1.head()
```

```python
bins = list(range(0, 3000, 300))

bins_label = [x / 300 for x in bins]
df1['price_label'] = pd.cut(df['price'], bins, right = False, labels = bins_label[: -1])
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1['price_label'].isnull().sum()
```

```python
df_product = df1[['product_id', 'purchased', 'cart', 'view']]
df_product
```

```python
df_product1 = df_product.groupby('product_id')['purchased', 'cart', 'view'].sum()
```

```python
df_product1
```

```python
df_product1['rating_value'] = (df_product1['purchased'] + df_product1['cart'] * 0.7 + df_product1['view'] * 0.1) / (df_product1['purchased'] + df_product1['cart'] + df_product1['view'])

df_product1
```

```python
df_pr = df_product1['rating_value']
```

```python
df2 = pd.merge(df1, df_pr, on = 'product_id', how = 'outer')
```

```python
df2
```

```python
df3 = df2[['event_type', 'product_id', 'user_id', 'price_label','rating_value']]

df3
```

```python
df3['price_label'].isnull().sum()
```

```python
df3.to_parquet('2019-Oct-rating.parquet')
```
